## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated dataplot a scatter plot.


Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
from matplotlib import pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.plot(x,y)
plt.xlabel("X label ")
plt.ylabel("Y label ")
plt.show
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 50])

plt.subplot(2,2,1)
plt.plot(x1,y1)

x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x2,y2)


x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3= np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x3,y3)


x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,4)
plt.plot(x4,y4)

plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

bar plot is chart represents the data in the form of bars ( in rectangular shapes) 


```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.subplot(1,2,1)
plt.bar(company,profit)

plt.subplot(1,2,2)
plt.barh(company,profit)

plt.show()

```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot visually summarizes the distribution of a dataset, depicting key statistics like
median, quartiles, and potential outliers. It provides a concise representation of variability and
central tendency, aiding in comparing and understanding data distributions, especially in
statistical analysis and identifying outliers in a dataset.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot((list([box1,box2])))
plt.show()
```
