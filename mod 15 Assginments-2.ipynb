we use while loop instead of for loop when the loop can be incremented or decremented depending on the user specific conidition , but not when the loop has to increment or decrement gradually 

```python
print("here for loop is used for consequent increment")
for i in range(10):
    print(i,end=" ")

print("\nhere while loop is used and increment is done on user specific condition ")

i=0
while i<10:
    print(i,end=" ")
    i=i+2

```

```python
for i in range(7):
    for j in range(i):
        for k in range(j):
            print('*',end=" ")
    
    print()
```
