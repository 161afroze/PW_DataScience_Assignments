## 3)  In what situations is nominal encoding preferred over one-hot encoding? Provide a practical example.

One-hot encoding is preferred over ordinal encoding when categorical variables lack inherent order, as in meal types like "breakfast," "lunch," "dinner." It ensures each category is treated independently, vital when no ordinal relationship should be assumed. Especially beneficial for datasets with numerous categories, one-hot encoding assigns each unique value its binary dimension, maintaining information integrity without introducing hierarchy. In restaurant reviews, one-hot encoding accurately represents meal types without implying any order, making it ideal for preserving categorical distinctions in machine learning models.
